```python
import numpy as np
import scipy.sparse as sp
import pandas as pd

def reverse_complement(seq) :
    rc_seq = ''
    
    for b in seq :
        if b == 'A' :
            rc_seq = 'T' + rc_seq
        elif b == 'C' :
            rc_seq = 'G' + rc_seq
        elif b == 'G' :
            rc_seq = 'C' + rc_seq
        elif b == 'T' :
            rc_seq = 'A' + rc_seq
    
    return rc_seq
```

```python

tissue_dfs = [
    apadb_df.join(tissue_df.set_index('gene_id')[['derti_cut_mode', 'supporting_reads', 'pas_info']], how='inner').reset_index()
    for tissue_df in tissue_dfs
]
```

```python
#Assign derti cluster to closest APADB annotation (by cut mode)

for tissue_ix in range(len(tissue_dfs)) :
    
    min_derti_site_dict = {}
    
    tissue_df = tissue_dfs[tissue_ix]
    
    for _, row in tissue_df.iterrows() :
        
        gene_id = row['gene_id']
        derti_id = str(row['derti_cut_mode']) + "_" + row['pas_info']
        site_dist = np.abs(row['cut_mode'] - row['derti_cut_mode'])
        
        if derti_id not in min_derti_site_dict or site_dist < min_derti_site_dict[derti_id]['site_dist'] :
            min_derti_site_dict[derti_id] = {
                'gene_id' : gene_id,
                'site_dist' : site_dist
            }
    
    keep_index = []
    i = 0
    for _, row in tissue_df.iterrows() :
        
        gene_id = row['gene_id']
        derti_id = str(row['derti_cut_mode']) + "_" + row['pas_info']
        
        if gene_id == min_derti_site_dict[derti_id]['gene_id'] :
            keep_index.append(i)
        
        i += 1
    
    tissue_dfs[tissue_ix] = tissue_df.iloc[keep_index].copy().reset_index(drop=True)
```

```python
#Remove wrongfully assigned CSEs

for tissue_ix in range(len(tissue_dfs)) :
    
    tissue_df = tissue_dfs[tissue_ix]
    
    keep_index = []
    i = 0
    for _, row in tissue_df.iterrows() :
        
        pas_info = row['pas_info'].split("|")[1]
        
        if pas_info == 'none' and row['pas'] in [-1, 3, 4] :
            keep_index.append(i)
        elif row['seq'][50:56] == pas_info :
            keep_index.append(i)
        
        i += 1
    
    tissue_dfs[tissue_ix] = tissue_df.iloc[keep_index].copy().reset_index(drop=True)
```
